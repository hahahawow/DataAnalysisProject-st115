```python
# just some rough exploratory data, these aren't final graphs I was just curious what it looked like

import pandas as pd
departmental_publications_data = pd.read_csv("Data/departmental_publications_date.csv")
departmental_publications_data.tail()
```

```python
import plotly.graph_objects as go

total_publications_count = departmental_publications_data.groupby('Date').size().reset_index(name='Publication Count')

# line plot
fig = go.Figure(data=go.Scatter(x=total_publications_count['Date'], y=total_publications_count['Publication Count'], mode='lines'))
fig.update_layout(title='Total Number of Publications Over Time', xaxis_title='Date', yaxis_title='Publication Count')

fig.show()
```

```python
import plotly.express as px

# grouping by department and date, count the number of publications
publications_count = departmental_publications_data.groupby(['Department', 'Date']).size().reset_index(name='Publication Count')

# creating line graph
fig = px.line(publications_count, x='Date', y='Publication Count', color='Department', title='Number of Publications Over Time by Department')

# adding a range slider
fig.update_layout(xaxis=dict(
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(count=5, label="5y", step="year", stepmode="backward"),
            dict(step="all")
        ])
    ),
    rangeslider=dict(visible=True),
    type="date"
))

fig.show()
```
